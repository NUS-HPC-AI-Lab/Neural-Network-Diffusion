```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import pickle
import os
```

```python
df = pd.DataFrame({
    'time': np.tile(np.arange(0, 1000, 2), 5),
    'value': np.concatenate([np.random.randn(500).cumsum() for _ in range(5)]),
    'random seed': np.repeat(['430', '431', '432', '433', '434'], 500)
})
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Liberation Sans', 'Arial', 'DejaVu Sans', 'Helvetica']
fig, ax1 = plt.subplots(figsize=(12, 9))

sns.lineplot(x='time', y='value', hue='random seed', data=df)

#ax1.set_ylim(0.0, 1.0)
ax1.tick_params(axis='both', which='major', labelsize=28)
ax1.set_ylabel("accuracy (%)", fontsize=28)
ax1.set_xlabel("diffusion steps", fontsize=28)
#ax1.legend(handles=[org_legend, gen_legend], loc='lower left', fontsize=28)
plt.legend(title="random seed", fontsize=22, title_fontsize=24, loc='lower right')

plt.tight_layout()
plt.savefig("draw_diffusion_process.pdf")
```
